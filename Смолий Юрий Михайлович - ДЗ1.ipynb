### 1. Что такое "несмещенная оценка"? Что это св-во означает? 
$$E[Q]=Q_0$$ 
"Несмещенная оценка" , это такая оценка, - математическое ожидание, которой, стремится к истенному значению оценки, для любого допустимого значения из области ее определения. 
На практике это означает , что если мы возмем несколько выборок из одного и того же распределения, посчитаем по каждой "оценку", то мат.ожидание "оценки" будет стремиться к истенному значению "оценки" , при устремлении кол-ва выборок к бесконечности (при N $\to\infty$).
Простым языком - несмещенность "оценки" говорит о правильном расчете данной "оценки".


### 2. Что такое "состоятельная оценка"? ЧТо это св-во означает? 
$$ Q(X)\to_{P_Q} Q_0 $$
$$ P( |Q(X) - Q_0| > e) \to 0 , \text{при любом e, при N}\to\infty$$

"состоятельная оценка", это такая оценка , которя для любого своего значения из области своего определения, сходится
по вероятности к своему истенному значению, при N $\to\infty$.
"состоятельность" - означает концентрацию оценки около истенного значения, при N $\to\infty$, т.е. истеное значение и ее оценка должны совпасть.


### 3. Что такое функция правдоподобия? Что она означает ?
$$ L(Q) = p(x_1,Q)*p(x_2,Q)*...*p(x_n,Q) $$
Функцией правдоподобия называют произведение значений плотностей вероятности для каждого наблюдения $ x_i $  (или произведение значений вероятности, для дискретного распределения). Функция правдоподобия по сути - это вероятность (при заданом параметре Q) получить именно такую реализацию выборки из закона распределения.

```python
import numpy as np
np.random.seed()

# Сгенерируем Q из равномерного распределения на [25,50]
# число тестов Q пока менять нельзя, со врмеменем можно расширить функции теста и добавить эту размерность

Число_тестов_Q = 1 
Правая_Q = np.random.uniform(25,50, size=[Число_тестов_Q])
print("нужно оценить границу Q для интервала [0, Q], истинное значение=",Правая_Q,'\n')

# Сгенерируем выборки из равномерного распредеоения [0, Q] размера n=10, 100, 1000, 10000

Размер_выбороки = [10, 100, 1000, 10000]
Выборки_на_0_Q = [list(np.random.uniform(0, Правая_Q, size=[размер1])) for размер1 in Размер_выбороки ]

# находим значения полученных оценок
# Оценка методом моментов Q = 2*Xср
# Оценка методом правдоподобия Q = max(Xi)

ММомент_ГраницаQ = [(Размер_выбороки[i], 2*np.mean(Выборки_на_0_Q[i]) ) for i in range(len(Выборки_на_0_Q)) ]
МПравдоп_ГраницаQ = [(Размер_выбороки[i], np.max(Выборки_на_0_Q[i]) ) for i in range(len(Выборки_на_0_Q)) ]

{print("выборка n=",ММомент_ГраницаQ[i][0], "М.Моментов оценка Q=", ММомент_ГраницаQ[i][1], 
       "  М.правд. оценка Q=", МПравдоп_ГраницаQ[i][1]) 
 for i in range(len(ММомент_ГраницаQ))}
print(' ')

# выводим отклонение полученых оценок от рпараметра Q. Что происходит с ростом N?

{print("выборка n=",ММомент_ГраницаQ[i][0], "М.Моментов отклонение dQ=", abs(ММомент_ГраницаQ[i][1]-Правая_Q), 
       "  М.Правдоп. отклонение dQ=", abs(МПравдоп_ГраницаQ[i][1]-Правая_Q)) 
 for i in range(len(ММомент_ГраницаQ))}
```

```python
import numpy as np # для генерации случайных величин и работы с массивами
import math # чтобы округлить число

np.random.seed() # фиксируем seed

# для оценки площади всей фигуры с дисперсией D_фигуры необходимо 80000 наблюдений
# для оценки площади сектора с дисперсией (1/4)*D_фигуры нам необходимо 20000 наблюдений в 4 р меньше.

n = 20000

# генерируем выборку из равномерного распределения, на участке x = [0, 1] 
# для оценки площади под функцией y(x)=x^2 

x = np.random.rand(n)

# ищем значение y(x)
y = [ x[i]**2 for i in range(n)]

# оцениваем площадь всей фигуры
S = 4*np.mean(y) 

print("Истинная площадь Фигуры=", 4/3)
print("Оценка площади Фигуры=", S)
print("Точность=", math.ceil(100*abs(S-4/3))/100  )
```
